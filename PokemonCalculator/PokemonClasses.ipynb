```python
import csv
import random
```

```python
class Moves():
    Name = ""
    Type = ""
    PP = 0
    Power = 0
    Acc = 0
    def __init__(self, Name, Type, PP, Power, Acc):
        self.Name = Name
        self.Type = Type
        self.PP = PP
        self.Power = Power
        self.Acc = Acc
```

```python
MoveNum = []
with open('PokemonGen1Moves.csv', 'r') as f:
    reader = csv.reader(f)
    for row in reader:
        MoveNum.append(Moves(row[0], row[1], row[2], row[3], row[4]))
MoveNum[1].Name
```

```python
class Pokemon():
    Name = " "
    Type1 = " "
    Type2 = " "
    HP = 0
    Attack = 0
    Defense = 0
    SpecialAtk = 0
    SpecialDef = 0
    Speed = 0
    Level = 50
    Ability = ""
    Move1 = ""
    Move2 = ""
    Move3 = ""
    Move4 = ""
    def __init__(self, Name, Type1, Type2, HP, Attack, Defense, SpecialAtk, SpecialDef, Speed, Ability):
        self.Name = Name
        self.Type1 = Type1
        self.Type2 = Type2
        self.HP = HP
        self.Attack = Attack
        self.Defense = Defense
        self.SpecialAtk = SpecialAtk
        self.SpecialDef = SpecialDef
        self.Speed = Speed
        self.Ability = Ability
```
