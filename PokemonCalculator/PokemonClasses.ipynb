```python
import csv
```

```python
class Moves():
    Name = ""
    Type = ""
    PP = 0
    Power = 0
    Acc = 0
    def __init__(self, Name, Type, PP, Power, Acc):
        self.Name = Name
        self.Type = Type
        self.PP = PP
        self.Power = Power
        self.Acc = Acc
```

```python
MoveNum = []

with open('PokemonGen1Moves.csv', 'r') as f:
    reader = csv.reader(f)
    for row in reader:
        MoveNum.append(Moves(row[0], row[1], row[2], row[3], row[4]))
MoveNum[1].Name
```

```python
class Pokemon():
    Name = " "
    Type1 = " "
    Type2 = " "
    HP = 0
    Attack = 0
    Defense = 0
    Special = 0
    Speed = 0
    def __init__(self, Name, Type1, Type2, HP, Attack, Defense, Special, Speed):
        self.Name = Name
        self.Type1 = Type1
        self.Type2 = Type2
        self.HP = HP
        self.Attack = Attack
        self.Defense = Defense
        self.Special = Special
        self.Speed = Speed
```

```python
PokemonNum = []

with open('FirstGenPokemon.csv', 'r') as f:
    reader = csv.reader(f)
    for row in reader:
        PokemonNum.append(Pokemon(row[1], row[2], row[3], row[4], row[5], row[6], row[7], row[8]))
PokemonNum[1].Type1
```
